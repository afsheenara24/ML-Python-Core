```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Features: [Area, Bedrooms, Age]
X = np.array([
    [1200, 2, 20],
    [1500, 3, 15],
    [1800, 3, 10],
    [2000, 4, 5],
    [2300, 4, 3]
])

# Target: House Price (in lakhs)
y = np.array([50, 65, 75, 85, 95])
```

```python
X_mean = np.mean(X, axis=0)
X_std = np.std(X, axis=0)

X_scaled = (X - X_mean) / X_std
```

```python
m = X_scaled.shape[0]
X_bias = np.c_[np.ones(m), X_scaled]
```

```python
theta = np.zeros(X_bias.shape[1])  # [bias, w1, w2, w3]
learning_rate = 0.01
iterations = 1000
```

```python
for _ in range(iterations):
    y_pred = X_bias.dot(theta)
    error = y_pred - y

    gradients = (1/m) * X_bias.T.dot(error)
    theta -= learning_rate * gradients
```

```python
print("Learned parameters:", theta)
```

```python
# Mean Squared Error
mse = np.mean((y - y_pred) ** 2)

# R-squared
ss_total = np.sum((y - np.mean(y)) ** 2)
ss_residual = np.sum((y - y_pred) ** 2)
r2 = 1 - (ss_residual / ss_total)

print("MSE:", mse)
print("R2 Score:", r2)
```

```python
plt.scatter(y, y_pred)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual vs Predicted Prices")
plt.show()
```
